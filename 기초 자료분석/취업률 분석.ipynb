#### * 대학 과정 고등교육기관 학제는 일반대학, 교육대학, 방송통신대학, 산업대학, 기술대학, 각종학교(대학과정), 사이버대학(대학과정), 원격대학형태의 평생교육시설(대학과정), 사내대학형태의 평생교육시설(대학과정)에 해당함															


## 전체 내용 출력

```python
import pandas as pd

job = pd.read_csv("Job.csv",encoding="cp949")

job.head(10)
```

## 소계열 개수 출력

```python
job["소계열"].count()
```

## 취업률이 가장 높은 전공(20명 이상)

```python
max_value = job[job["졸업자_계"] >= 20]["취업률_계"].max()

job[job["취업률_계"]==max_value]["소계열"]
```

## 대계열별, ' 졸업자_계', '졸업자_남'	'졸업자_여'	,  '취업률_계', 	'취업률_남', 	'취업률_여' 을 표로 만들고 이를 'job_great.csv' 로 저장

```python
job_sub = job.copy()

def fill(series):
    series = list(series)
    for i in range(len(series)):
        if series[i] == "-":
            series[i] = 0
    return series

for i in job.columns:
    job_sub[i] = pd.Series(fill(job[i]))
```

```python
job_great_sub = job_sub.iloc[:,[0,3,4,5,6,7,8,9,10,11]]
job_great_sub
```

## 데이터 type 확인

```python
columns = list(job_great_sub.columns)

job_great_sub = job_great_sub.astype({columns[1] : int,columns[2] : int,columns[3] : int,columns[7] : int,columns[8] : int,columns[9] : int})
job_great_sub = job_great_sub.astype({columns[4] : float,columns[5] : float,columns[6] : float})
job_great_sub.info()

```
