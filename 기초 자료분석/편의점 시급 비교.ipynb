## 전체 데이터 출력

```python
import pandas as pd

alba = pd.read_csv("alba.csv", encoding="utf-8")

alba.head(10)
```

## 기술 통계 정보 요약하여 출력

```python
alba.describe()
```

## 최고시급, 최저시급 출력

```python
max_value = alba['hourly_wage'].max()

alba[alba["hourly_wage"] == max_value]
```

## 세븐일레븐(7/11) 편의점 시급의 평균

```python
seven = alba.loc[alba["name"] == "7/11"]["hourly_wage"].mean()
seven
```

## 강남구 편의점 시급의 평균

```python
gangnam = alba.loc[alba["area1"] == "강남구"]["hourly_wage"].mean()
gangnam
```

## 구별 점포 수를 새로운 데이터 프레임에 가공, 출력

```python
alba_sub = pd.DataFrame({
    "구" : ['강남구', '강동구', '강북구', '강서구', '관악구', ' 광진구', ' 구로구', ' 금천구', ' 노원구', ' 도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', ' 은평구', '종로구', '중구', '중랑구'],
    "점포 수" : [i for i in alba.groupby(["area1"])["area"].count()],
    "시급평균" : [i for i in alba.groupby(["area1"])["hourly_wage"].mean()],
    "시급최고" : [i for i in alba.groupby(["area1"])["hourly_wage"].max()],
    "시급최고" : [i for i in alba.groupby(["area1"])["hourly_wage"].min()]
})
```
