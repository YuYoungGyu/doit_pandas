## 전체 데이터 출력

```python
import pandas as pd

data = pd.read_excel(f"data.xlsx",encoding="cp949")


display(data)
```

## 그래프 그리기

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

data_graph = pd.DataFrame({
    "대학원" : [i for i in data.loc[9:15,"평균소득"]],
    "학부" : [i for i in data.loc[1:7,"평균소득"]]
}, index = data.loc[1:7,"구분"])

ax = data_graph.plot(kind = "bar",rot = 0,width = 0.8,color = ["b","k"])

plt.xlabel("계열")
plt.ylabel("월\n소\n득\n평\n균\n(천원)", labelpad = 10, rotation = 0)

display(ax)
```
